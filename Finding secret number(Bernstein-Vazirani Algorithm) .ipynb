```python
from qiskit import*
```

```python
from qiskit.visualization import plot_histogram
```

```python
%matplotlib inline
```

```python
secret_num = '1000100010'
```

```python
qc = QuantumCircuit(len(secret_num)+1, len(secret_num))
```

```python
qc.h(range(len(secret_num)))
qc.x(len(secret_num))
qc.h(len(secret_num))
qc.barrier()
qc.draw('mpl')
```

```python
for iz, val in enumerate(reversed(secret_num)):
    if(val=='1'):
        qc.cx(iz, len(secret_num))
qc.barrier()
qc.draw('mpl')
```

```python
qc.h(range(len(secret_num)))
qc.barrier()
qc.measure((range(len(secret_num))),(range(len(secret_num))))
qc.draw('mpl')
```

```python
job = execute(qc, backend=Aer.get_backend('qasm_simulator'), shots=1024)
counts = job.result().get_counts()
print(counts)
```
