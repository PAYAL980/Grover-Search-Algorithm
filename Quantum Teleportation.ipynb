## Quantum Teleportation 

```python
from qiskit import*
```

```python
%matplotlib inline
```

```python
ckt = QuantumCircuit(3,3)
```

```python
ckt.draw('mpl')
```

Transferring state $|1\rangle$ from qubit0 to qubit3 by using qubit2

```python
ckt.x(0)
ckt.barrier()
ckt.h(1)
ckt.cx(1,2)
ckt.cx(0,1)
ckt.h(0)
ckt.barrier()
ckt.measure([0,1], [0,1])
ckt.draw('mpl')
```

```python
ckt.barrier()
ckt.cx(1,2)
ckt.cz(0,2)
ckt.draw('mpl')
```

```python
ckt.measure(2,2)
job = execute(ckt, backend = Aer.get_backend('qasm_simulator'), shots=1024)
counts = job.result().get_counts()
from qiskit.visualization import plot_histogram
plot_histogram(counts)
```

Transferring state $|0\rangle$ from qubit0 to qubit3 by using qubit2

```python
ckt1 = QuantumCircuit(3,3)
ckt1.h(1)
ckt1.cx(1,2)
ckt1.cx(0,1)
ckt1.h(0)
ckt1.barrier()
ckt1.measure([0,1], [0,1])
ckt1.barrier()
ckt1.cx(1,2)
ckt1.cz(0,2)
ckt1.measure(2,2)
job = execute(ckt1, backend = Aer.get_backend('qasm_simulator'), shots=1024)
counts = job.result().get_counts()
plot_histogram(counts)
```

```python
ckt1.draw('mpl')
```

Transporting state $\frac{|0\rangle + |1\rangle}{\sqrt(2)}$ from qubit0 to qubit2

```python
ckt2 = QuantumCircuit(3,3)
ckt2.h(0)
ckt2.barrier()
ckt2.draw('mpl')
```

```python
ckt2.h(1)
ckt2.cx(1,2)
ckt2.cx(0,1)
ckt2.h(0)
ckt2.barrier()
ckt2.measure([0,1],[0,1])
ckt2.barrier()
ckt2.cx(1,2)
ckt2.cz(0,2)
ckt2.measure(2,2)
ckt2.draw('mpl')
```

```python
job2 = execute(ckt2, backend=Aer.get_backend('qasm_simulator'), shots=1024)
counts = job2.result().get_counts()
plot_histogram(counts)
```
